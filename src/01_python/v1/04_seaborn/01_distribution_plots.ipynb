### Distribution Plots

````python
import seaborn as sns
%matplotlib inline

# built-in DataFrame
tips = sns.load_dataset('tips')
tips.head(3)
````

## **displot**

`sns.displot()`: figure-level function for plotting distribution visualizations.

**args**:
- ```data=None```: Dataset for plotting.
- ```x=None```: Data variable for x-axis.
- ```y=None```: Data variable for y-axis.
- ```aspect=1```: Aspect ratio of each facet.
- ```binrange=None```: Range for bin edges.
- ```bins='auto'```: Data intervals. Each bin represents a specific interval of values, and the height of the bar represents the frequency (number of data points) that fall into that interval. Bars in histograms are called 'bins' ;)
- ```binwidth=None```: Width of each bin (alternative to bins).
- ```col=None```: Variable in data to map to subplot columns.
- ```col_order=None```: Order for subplot columns.
- ```col_wrap=None```: Wrap column variable at this width.
- ```color=None```: Color for all elements.
- ```common_bins=True```: Whether to use the same bin edges for all facets.
- ```common_norm=True```: Whether to normalize histograms.
- ```cumulative=False```: Whether to plot a cumulative distribution.
- ```discrete=None```: Whether data should be treated as discrete.
- ```element='bars'```: Approach to represent data in histogram.
- ```facet_kws=None```: Keyword arguments for the underlying faceting function.
- ```fill=True```: Whether to fill the histogram bars.
- ```height=5```: Height of each facet in inches.
- ```hue=None```: Variable in data to map to histogram color.
- ```hue_norm=None```: Pair of numbers to scale the hue variable.
- ```hue_order=None```: Order for the levels of the hue variable.
- ```kind='hist'```: Kind of plot to draw ('hist', 'kde', or 'ecdf').
- ```kde=False```: Whether to plot a gaussian kernel density estimate.
- ```kde_kws=None```: Keyword arguments for kdeplot().
- ```kde_norm=True```: Whether to normalize the KDE.
- ```legend=True```: Whether to add a legend.
- ```line_kws=None```: Keyword arguments for line plots (for element='step' or 'poly').
- ```log_scale=None```: Base of logarithm for x, y axes.
- ```multiple='layer'```: Approach to resolve multiple elements when semantic mapping creates subsets.
- ```palette=None```: Colors to use for different levels of hue.
- ```pmax=None```: Maximum proportion of dataset for large bin count.
- ```pthresh=None```: Proportional threshold for small histograms to reassign to nearest bin.
- ```rug=False```: Whether to add rug plots.
- ```rug_kws=None```: Keyword arguments for rugplot().
- ```row=None```: Variable in data to map to subplot rows.
- ```row_order=None```: Order for subplot rows.
- ```shrink=1```: Scale down the size of the plot elements.
- ```stat='count'```: Aggregate statistic to compute in each bin.
- ```thresh=0```: Threshold for small histograms to reassign to nearest bin.
- ```weights=None```: Weight variable to reweight statistics in each bin.
- ```**kwargs```: Other keyword arguments are passed to the underlying plotting function based on kind.

````python
# univariate distribution (focuses on visualizing one variable at a time)
sns.displot(tips["total_bill"], bins=30, kde=True)
````

## **jointplot**

````python
# multivariate distribution
sns.jointplot(x="total_bill", y="tip", data=tips)
````

````python
# kind='hex' -> if hexagon has certain number of points in it, it gets
#   darker: if it has less points, it gets lighter.
sns.jointplot(x='total_bill',y='tip',data=tips, kind='hex')
````

````python
# includes regression line: shows a linear fit of the scatter data.
sns.jointplot(x='total_bill',y='tip',data=tips, kind='reg')
````

````python
sns.set_style("whitegrid") # force VSCode to use white grid
# shows the density where the points match up most
sns.jointplot(x='total_bill',y='tip',data=tips, kind='kde', cmap="Blues", fill=True)
````

## **pairplot**

````python
# - pairplot() will do the jointplot() for every possible combination in the
# numerical columns from the given DataFrame
# - Nice way to quickly visualise our data
sns.pairplot(tips)
````

````python
# hue: use categorical columns (e.g.: sex: male or female) -> will color the
# datapoints based on the hue column
sns.pairplot(tips, hue='sex', palette='coolwarm')
````

## **rugplot**

````python
# same concept as histogram but instead of bins, but shows a dashmark
# for every single point along the distribution line
sns.rugplot(tips['total_bill'])
````

## **kdeplot**

````python
# just draws the kde
sns.kdeplot(tips['total_bill'])
````
