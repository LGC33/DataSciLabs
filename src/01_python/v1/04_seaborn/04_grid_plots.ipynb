# Grid Plots

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')
iris.head()
```

## **PairGrid**

`seaborn.PairGrid()`: Subplot grid for plotting pairwise relationships in a dataset.

**Parameters**:

- `data`: Tidy (long-form) dataframe where each column is a variable and each row is an observation.
- `hue=None`: Variable in `data` to map plot aspects to different colors. This variable will be excluded from the default x and y variables.
- `vars=None`: Variables within `data` to use, otherwise use every column with a numeric datatype.
- `x_vars=None`, `y_vars=None`: Variables within `data` to use separately for the rows and columns of the figure; i.e. to make a non-square plot.
- `hue_order=None`: Order for the levels of the hue variable in the palette.
- `palette=None`: Set of colors for mapping the hue variable. If a dict, keys should be values in the hue variable.
- `hue_kws=None`: Other keyword arguments to insert into the plotting call to let other plot attributes vary across levels of the hue variable.
- `corner=False`: If True, don’t add axes to the upper (off-diagonal) triangle of the grid, making this a “corner” plot.
- `diag_sharey=True`: Whether to share the y-axis across a row of diagonals.
- `height=2.5`: Height (in inches) of each facet.
- `aspect=1`: Aspect * height gives the width (in inches) of each facet.
- `layout_pad=0.5`: Padding between axes; passed to `fig.tight_layout`.
- `despine=True`: Remove the top and right spines from the plots.
- `dropna=False`: Drop missing values from the data before plotting.

```python
# similar to pairplot but can be customized and gives more control
g = sns.PairGrid(iris, height=1.75)
g.map(plt.scatter) # map scatter to PairGrid
```

```python
g = sns.PairGrid(iris, height=1.75)
g.map_diag(sns.histplot) # map diagonal to histplot (distplot is deprecated)
g.map_upper(plt.scatter) # map upper triangle (upper-right) to scatter plots
g.map_lower(sns.kdeplot) # map lower triangle (lowwer-left) to KDE plots
```

## **FacetGrid**

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.histplot, 'total_bill')
```

```python
import matplotlib.pyplot as plt
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(plt.scatter, 'total_bill', 'tip')
```
