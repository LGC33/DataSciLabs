````python
import matplotlib.pyplot as plt
import numpy as np

# only for Jupyter notebook
%matplotlib inline

x = np.linspace(0,5,11)
# [ 0.0 0.5 1.0 1.5 2.0 2.5 3.0 3.5 4.0 4.5 5.0 ]
y = x ** 2
# [ 0.0 0.25 1.0 2.25 4.0 6.25 9.0 12.25 16.0 20.25 25.0 ]
````

sign.: ```add_axes([left, bottom, width, height])```

args:
- ```left```: This is the position of the left side of the axes. In this case, it's 10% from the left side of the canvas.
- ```bottom```: This is the position of the bottom of the axes. Here, it's also 10% from the bottom of the canvas.
- ```width```: The width of the axes. 0.8 means the axes will take up 80% of the figure canvas's width.
- ```height```: The height of the axes. Similarly, 0.8 means the axes will take up 80% of the figure canvas's height.

returns:
- ```fig``` : Figure. The created figure object.
- ```ax``` : Axes object or array of Axes objects: can be either a single Axes object or an array of Axes objects if more than one subplot was created.

## Functional Method


````python
plt.plot(x, y)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")
````

````python
plt.subplot(1, 2, 1)  # rows, columns, plot number
plt.plot(x, y, "r")  # 'r': color red
plt.subplot(1, 2, 2)
plt.plot(y, x, "b")
````

## Object Oriented Method

````python
fig = plt.figure()
# @dev: left & bottom only work when there are 2+ axes. If only one, values will be set by default
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # left, bottom, witdh, high
axes.plot(x, y)
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Set Title")
````

````python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes1.set_title("Larger Plot")
axes2.plot(y, x)
axes2.set_title("Smaller Plot")
````

sign.: ```plt.subplots(nrows=1, ncols=1, sharex=False, sharey=False, squeeze=True, subplot_kw=None, gridspec_kw=None, **fig_kw)```

info: creates both a figure and a set of subplots in a single step. This is a powerful tool for creating multiple plots within the same figure.

args:
- ```nrows```, ncols : int, optional, default: 1.
    Number of rows/columns of the subplot grid.
- ```sharex```, sharey : bool or {'none', 'all', 'row', 'col'}, default: False
    Controls sharing of properties among x (sharex) or y (sharey) axes:
    - True or 'all': x or y axis will be shared among all subplots.
    - False or 'none': each subplot has its own independent x or y axis.
    - 'row': each subplot row will share an x or y axis.
    - 'col': each subplot column will share an x or y axis.
- ```squeeze``` : bool, optional, default: True
    If True, extra dimensions are squeezed out from the returned Axes object(s):
    - if only one subplot is constructed (nrows=1 or ncols=1), the resulting single Axes object is returned as a scalar.
    - for Nx1 or 1xM subplots, the returned object is a 1D numpy array.
    - for NxM subplots with N>1 and M>1, the returned object is a 2D numpy array.
- ```subplot_kw``` : dict, optional
    Dict with keywords passed to the add_subplot call used to create each subplot.
- ```gridspec_kw``` : dict, optional
    Dict with keywords passed to the GridSpec constructor used to create the grid the subplots are placed on.
- **```fig_kw``` :
    All additional keyword arguments are passed to the plt.figure call upon subplot creation.

````python
fig, axes = plt.subplots(nrows=1, ncols=2)

# for current_ax in axes:
#     current_ax.plot(x,y)

axes[0].plot(x, y)
axes[0].set_title("1st plot")

axes[1].plot(y, x)
axes[1].set_title("2nd plot")

plt.tight_layout()  # to avoid overlaping
````

## Figure Size and DPI

sign.: ```plt.figure(num=None, figsize=None, dpi=None, facecolor=None, edgecolor=None, frameon=True, FigureClass=<class 'matplotlib.figure.Figure'>, clear=False, **kwargs)```

args:
- ```num``` : integer or string, optional, default: None. A unique identifier for the figure.
    If a figure with that identifier already exists, it makes it the current figure.
- ```figsize``` : (float, float), optional, default: None. Width, height in inches. Specifies the size of the figure.
- ```dpi``` : integer, optional, default: None. Resolution of the figure in dots-per-inch.
- ```facecolor``` : color, optional, default: None. The background color of the figure.
- ```edgecolor``` : color, optional, default: None. The border color of the figure.
- ```frameon``` : bool, optional, default: True. If False, suppress drawing the figure background patch.
- ```FigureClass``` : subclass of matplotlib.figure.Figure. Optionally use a custom Figure instance.
- ```clear``` : bool, optional, default: False. If True and the figure already exists, then it's cleared.
- ```**kwargs``` : All additional keyword arguments are passed to the matplotlib.figure.Figure constructor.

returns:
- ```figure``` : Figure. The created or selected figure.


````python
fig = plt.figure(figsize=(8, 2))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
````

````python
# plt.subplots() creates a 1x1 grid (i.e., one subplot) if you don't specify the
# nrows and ncols arguments.
fig, axes = plt.subplots(figsize=(8, 2))
axes.plot(x,y)
````

````python
fig, axes = plt.subplots(nrows=2,ncols=1, figsize=(8, 2))
axes[0].plot(x,y)
````

````python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 2))
axes[0].plot(x, y)
axes[1].plot(y, x)
plt.tight_layout()
````

sign.: ```plt.plot([x], y, [fmt], *, data=None, **kwargs)```


info: allows to plot lines and markers on the current axes.
  
args:
- ```[x]```: Sequence of x-values.
- ```y```: Sequence of y-values.
- ```[fmt]```: Format string (color, marker, linestyle).
- ```data```: Input data structure.
- ```**kwargs```: Additional keyword arguments.

Some of the frequently used keyword arguments (**kwargs) include:

- ```color (or c)```: Color of the line/marker.
- ```linewidth (or lw)```: Width of the line.
- ```linestyle (or ls)```: Style of the line ('-', ':', '-.', etc.).
- ```marker```: Marker style ('o', 's', '^', etc.).
- ```markersize (or ms)```: Size of the marker.
- ```markerfacecolor (or mfc)```: Color of the marker face.
- ```markeredgewidth (or mew)```: Width of the marker edge.
- ```markeredgecolor (or mec)```: Color of the marker edge.
- ```label```: Label for the line/marker (for legends).
- ```alpha```: The alpha transparency level (between 0 and 1). 1 is fully opaque, 0 is fully transparent.
- ```antialiased (or aa)```: Use antialiased rendering.
- ```dashes```: Dash sequence for line.
- ```drawstyle (or ds)```: Drawing style ('default', 'steps', etc.).
- ```markevery```: Mark a subset of data points using slice notation.
- ```solid_capstyle```: Cap style for solid lines.
- ```solid_joinstyle```: Join style for solid lines.
- ```visible```: Whether the rendered line/marker is visible (True/False).
- ```zorder```: Set the zorder for rendering.
- ```rasterized```: Rasterize the line when saving to raster-format images.


````python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='#FF8C00',
        linewidth=1,
        alpha=.7,
        linestyle='-',
        marker='o', 
        markersize=10, markerfacecolor='yellow', markeredgewidth=3,
        markeredgecolor='green')
# linestyle: - -- -. : steps
# marker: o + * 1
````

````python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='purple', lw=2, ls='--')
ax.set_xlim([0,1]) # lower & upper bounds in x-axis
ax.set_ylim([0,2]) # lower & upper bounds in y-axis
````
