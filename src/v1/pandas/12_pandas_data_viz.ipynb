# Pandas built-in data visualization

```python
import numpy as np
import pandas as pd
import seaborn as sns # will make the charts nicer
%matplotlib inline

# apply styles
sns.set_style("darkgrid") 

# load filez
df1 = pd.read_csv("./filez/df1", index_col=0)
df2 = pd.read_csv("./filez/df2")
```

```python
df1.head()
```

```python
df2.head()
```

```python
# Method 1
df1['A'].hist(bins=30)
```

```python
# Method 2
df1['A'].plot(kind='hist', bins=10)
```

```python
# Method 3
df1['A'].plot.hist()
```

```python
df2.plot.area(alpha=0.4)
```

```python
df2.plot.bar(stacked=True)
```

```python
# **** PENDING ************
df1 = df1.reset_index()
df1.plot.line(x=df1.index, y='B')
# df1.plot.line(x=df1.index, y='B', figsize=(12,3), lw=1)
```

```python
# shows A, B with color as C (so kind of 3D-chart)
df1.plot.scatter(x='A',y='B', c='C', cmap='coolwarm')
```

```python
# ************** REVIEW WARNING
# like before, but now the size indicates how large is C
df1.plot.scatter(x='A',y='B', s=df1['C']*100)
```

```python
df2.plot.box()
```
