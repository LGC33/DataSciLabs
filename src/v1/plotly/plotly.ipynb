```python
# libs
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.DataFrame(np.random.randn(100,4), columns='A B C D'.split())
df.head(4)
```

```python
df2 = pd.DataFrame({'Category': ['A','B','C'], 'Values': [32, 43, 50]})
df2
```

```python
# Non-interactive line plot
df.plot()
```

## **line**

```python
# Interactive line plot
fig = px.line(df)
fig.show()
```

## **scatter**

```python
# interactive scatter plot (only for values 'A' and 'B')
fig = px.scatter(df, x='A', y='B')
fig.update_traces(marker=dict(size=10)) # increase markers size
fig.show()
```

## **bar**

```python
# interactive bar plot
fig = px.bar(df2, x='Category', y='Values')
fig.show()
```

## **bar + count**

```python
# count plot (since all are even distributed, all are 100)
fig = px.bar(df.count())
fig.show()
```

## **bar + sum**

```python
fig = px.bar(df.sum())
fig.show()
```

## **box**

```python
fig = px.box(df)
fig.show()
```

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5], 'y':[10,20,30,20,10],'z': [500,400,300,200,100]})
# this DataFrame doesn't work with surface plots
```

```python
# Creates 5 evenly spaced numbers between the 'x' interval: from 1 (min) to 5 (max)
x = np.linspace(df3['x'].min(), df3['x'].max(), 5)
x
```

```python
# Creates 5 evenly spaced numbers between the 'y' interval: from 10 (min) to 30 (max)
y = np.linspace(df3['y'].min(), df3['y'].max(), 5)
y
```

```python
# Creating meshgrid for x and y
x, y = np.meshgrid(x, y)
x
```

```python
# Creating a function for z based on x and y
# Here you need to define how 'z' relates to 'x' and 'y'
# For example, if z = x^2 - y^2
z = x**2 - y**2
z
```

```python
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
# Update the layout
fig.update_layout(
    autosize=False,
    width=600,
    height=400,
    margin=dict(l=65, r=50, b=65, t=90)
)
fig.show()
```

```python
df4 = pd.DataFrame({'x':[0,1], 'y':[0,1]})
df4
```

```python
z = np.array([[5,10],[3,9]])
z
```

```python
fig = go.Figure(data=[go.Surface(x=df4['x'], y=df4['y'], z=z)])

# Update the layout
fig.update_layout(
    autosize=False,
    width=600,
    height=400,
    margin=dict(l=65, r=50, b=65, t=90)
)

fig.show()
```
