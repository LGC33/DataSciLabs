# Categorical Plots

````python
import seaborn as sns
import numpy as np
%matplotlib inline

# built-in DataFrame
tips = sns.load_dataset('tips')
tips.head(5)
````

## **barplot()**
Represents an estimate of central tendency for a numeric variable with the height of each rectangle

````python
sns.barplot(x="sex", y="total_bill", data=tips)
````

````python
# show what the std deviation of the total bill column is per category
sns.barplot(x="sex", y="total_bill", data=tips, estimator=np.std)
````

## **countplot()**

Shows the counts of observations in each categorical bin using bars.

````python
# same as barplot() except the estimator is explicitly counting the number
# of occurrencies, that's why we only set the x value (like pandas.count)
sns.countplot(x="sex", data=tips)
````

## **boxplot()**

Represents data distribution using quartiles, displaying the spread and skewness of continuous data for distinct categories. It can also indicate outliers.

````python
# shows the quartiles of total bill per day
# the dots are outliers
sns.boxplot(
    x="day",
    y="total_bill",
    data=tips,
    palette="rainbow",
    hue="day"
)
````

````python
# shows the quartiles of total bill per day and smoker
sns.boxplot(
    x="day",
    y="total_bill",
    data=tips,
    palette="rainbow",
    hue="smoker"
)
# e.g. conclusion: on Fridays, non-smokers tend to give a higher tip
````

## **violinplot()**

Combines aspects of a box plot and a KDE plot. It provides a visualization of the distribution of the data, its probability density, and its cumulative distribution.

````python
# same arguments as boxplot()
# harder to read, but gives more info
sns.violinplot(
    x="day",
    y="total_bill",
    data=tips,
    palette="rainbow",
    hue="sex",
    split=True
)
````

## **stripplot()**

Represents a scatter plot of points where one variable is categorical.

`sns.stripplot()`: displays a scatter plot based on categorical data.

args:
- ```x=None```: Input data variable for the x-axis.
- ```y=None```: Input data variable for the y-axis.
- ```hue=None```: Grouping variable that produces points with different colors.
- ```data=None```: DataFrame when x and y are variable names.
- ```order=None```: Order to plot the categorical levels in.
- ```hue_order=None```: Order to plot the hue levels in.
- ```jitter=True```: Amount of jitter (only along the categorical axis) or bool. If True, jitter is determined automatically. To show the density a bit better.
- ```dodge=False```: Separate the strips for different hue levels along the categorical axis.
- ```orient=None```: Orientation of the plot (v or h).
- ```color=None```: Color for all elements.
- ```palette=None```: Palette to use for different levels of hue.
- ```size=5```: Radius of the markers in points.
- ```edgecolor='gray'```: Color of the marker edges.
- ```linewidth=0```: Width of the marker edges in points.
- ```ax=None```: Pre-existing axes for the plot. Otherwise, a new one is created.
- ```**kwargs```: Other keyword arguments are passed to matplotlib.axes.Axes.scatter().


````python
sns.stripplot(
    x='day',
    y='total_bill',
    data=tips,
    palette='rainbow',
    hue='day',
)
````

````python
sns.stripplot(
    x='day',
    y='total_bill',
    data=tips,
    palette='rainbow',
    hue='sex',
    dodge=True
)
````

## **swarmplot()**

Similar to a strip plot, but the points are adjusted (only along the categorical axis) so they don't overlap.

````python
# combination of strip and violin plots
# swarmplot() is a categorical scatter plot that positions each point of the
# categorical variable so that they don't overlap. This gives a better
# representation of the distribution of values, although it doesn't scale well
# to large numbers of observations.
sns.swarmplot(
    x="day",
    y="total_bill",
    data=tips,
    palette="rainbow",
    hue="sex",
)
````

````python
# more for exploratory process, but hard to read
sns.violinplot(
    x="day", y="total_bill", data=tips, palette="rainbow", hue="sex", split=True
)
sns.swarmplot(
    x="day", y="total_bill", data=tips, palette="rainbow", hue="sex", color="black"
)
````

## **catplot()**

This is a higher-level interface to the above plots. It provides a way to draw categorical plots onto a FacetGrid. The kind parameter in catplot allows you to choose from any of the above plot types (e.g., kind="box", kind="violin", etc.).

````python
# factorplot() renamed to catplot()
# can show different plots depending on the kind, but it's generally better
# to use the specific plot function rather than this generic one
sns.catplot(
    x="day", y="total_bill", data=tips, palette="rainbow", hue="sex", kind="bar"
)
````
