# Regression Plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

## **lmplot**

Scatters plots with corresponding regression lines for examining linear relationships between two quantitative variables. It also allows for easy grouping of data points by 1+ categorical variables using color, rows, and columns, facilitating comparison across different categories. This function is particularly useful for exploratory data analysis when you need to investigate potential linear trends in the context of a larger multidimensional dataset.

`seaborn.lmplot()`: Combines regplot() and FacetGrid for plotting regression model fits across a dataset.

**Parameters**:

- `data=None`: Tidy dataframe for plotting.
- `x=None, y=None`: Variables for x and y axes.
- `hue=None`: Variable for color encoding.
- `col=None, row=None`: Variables for faceting.
- `palette=None`: Color palette.
- `col_wrap=None`: Wrap columns at this width.
- `height=5`: Height of each facet, in inches.
- `aspect=1`: Aspect ratio of each facet.
- `markers='o'`: Marker style.
- `sharex=None, sharey=None`: Share x or y axis across facets.
- `hue_order=None, col_order=None, row_order=None`: Order to plot the levels in a hue, col, or row.
- `legend=True`: Display legend.
- `x_estimator=None`: Function for x-axis aggregation.
- `x_bins=None`: Binning for x values.
- `x_ci='ci'`: Confidence interval for x estimator.
- `scatter=True`: Plot scatter points.
- `fit_reg=True`: Fit regression line.
- `ci=95`: Confidence interval for regression.
- `n_boot=1000`: Bootstrap resamples for CI.
- `units=None`: Sampling units for bootstrapping.
- `seed=None`: Seed for random number generator.
- `order=1`: Order of polynomial for regression.
- `logistic=False, lowess=False, robust=False, logx=False`: Regression options.
- `x_partial=None, y_partial=None`: Variables for partial regression.
- `truncate=True`: Truncate regression line.
- `x_jitter=None, y_jitter=None`: Jitter for x, y points.
- `scatter_kws=None, line_kws=None`: Additional kwargs for scatter and line plots.
- `facet_kws=None`: Kwargs for FacetGrid.

```python
sns.lmplot(data=tips, x="total_bill", y="tip")
```

```python
# shows 2 scatter plots and 2 linear fits: in this case, males & females have a similar linear fit
sns.lmplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex",
    markers=["o", "v"],
    scatter_kws={"s": 80},  # call to matplotlib directly to make the markers bigger
)
```

```python
sns.lmplot(
    data=tips,
    x="total_bill",
    y="tip",
    col="sex", # separate by columns instead of colours (like hue)
    row="time",
    height=3.5,  # New height
    aspect=1     # Keep the original aspect ratio
)
# hue can also be used like col & row
# to reduce the chart size proportionally, use height and aspect
```
